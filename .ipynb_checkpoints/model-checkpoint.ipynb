```python
import io
import ipywidgets as widgets
from IPython.display import display, Image as IPImage
import cv2
import numpy as np
import tensorflow as tf
from PIL import Image

# Load your CNN model
model = tf.keras.models.load_model('model')

# Define function to handle file upload and prediction
def handle_upload(change):
    uploaded_file = change['new'][0]
    content = uploaded_file['content']
    
    # Check if metadata exists
    if 'metadata' in uploaded_file:
        filename = uploaded_file['metadata'].get('name', 'unknown.jpg')
    else:
        filename = 'unknown.jpg'
    
    # Check file extension
    if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
        # Decode image content
        image = Image.open(io.BytesIO(content))
        # Convert to numpy array
        test_img = np.array(image)
        # Resize image to (256, 256)
        test_img = cv2.resize(test_img, (256, 256))
        # Reshape image for prediction
        pred_img = test_img.reshape((1, 256, 256, 3))
        
        # Display the image
        display(IPImage(data=content))
        
        # Make prediction
        prediction = model.predict(pred_img)
        if prediction[0][0] == 1:
            print('Prediction: Dog')
        elif prediction[0][0] == 0:
            print('Prediction: Cat')
        else:
            print("I can't predict")
    else:
        print('Please upload a JPG, JPEG, or PNG file.')

# Create file upload widget
uploader = widgets.FileUpload(accept='.jpg, .jpeg, .png', multiple=False)
uploader.observe(handle_upload, names='value')

# Display the file upload widget
display(uploader)
```
